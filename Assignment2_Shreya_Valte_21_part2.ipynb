# Cleaning of Data

1.remove Emojis


```python
df = df.astype(str).apply(lambda x: x.str.encode('ascii', 'ignore').str.decode('ascii'))
df
```

```python
df=df.drop(['Unnamed: 0'], axis=1)  
df
```

2.Remove Username

```python
df['Text'] = df['Text'].str.replace(r'\s*@\w+', '', regex=True)
df['Text'] = df['Text'].str.replace(r'\s*\B@\w+', '', regex=True)
df['Text'] = df['Text'].str.replace(r'\s*@\S+', '', regex=True)
df['Text'] = df['Text'].str.replace(r'\s*@\S+\b', '', regex=True)
```

```python
df.tail(40)
```

3.Remove Panctuations

```python
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 
# Removing punctuations in string
# Using loop + punctuation string
def remove_punct(text):
    for s in text:
        if s in punc:
            text = text.replace(s, "")
    return text
```

```python
df['Text']=df['Text'].apply(remove_punct)
df
```

Converting a text to a lower case

```python
df['Text'] = df['Text'].str.lower()
df
```

# Creating a labelled column

#Keywords present in list are labelled by 1 else 0


```python
df['label'] = df.Text.str.extract('(aap|bsp|bjp|sp|congress)')
df['label'] = df['label'].replace(['aap','bsp','bjp','sp','congress'],'1')
df['label'] = df['label'].fillna(0)
```
